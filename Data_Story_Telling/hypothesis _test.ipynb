```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway
```

```python
df = pd.read_csv( "cleaned_data.csv")
```

```python
df.head()
```

```python
df['Region'].unique()
```

```python
east_sales = df[df['Region'] == 'East']['Sales']
west_sales = df[df['Region'] == 'West']['Sales']
north_sales = df[df['Region'] == 'North']['Sales']
south_sales = df[df['Region'] == 'South']['Sales']
```

```python
f_stat, p_value = f_oneway(east_sales, west_sales, north_sales, south_sales)

print("F-statistic:", f_stat)
print("P-value:", p_value)
```

```python
if p_value < 0.05:
    print("Reject the Null Hypothesis")
    print("There is a statistically significant difference in sales between regions.")
else:
    print("Fail to Reject the Null Hypothesis")
    print("There is no statistically significant difference in sales between regions.")
```

```python

'''We performed a One-Way ANOVA test to determine whether there is a significant difference in sales across different regions (East, West, North, South).'''
''
Null Hypothesis (H0):
There is no significant difference in sales between regions.

Alternative Hypothesis (H1):
There is a significant difference in sales between regions.
```

```python
# Conclusion
Based on the ANOVA test results, the p-value is greater than 0.05.

Therefore, we fail to reject the null hypothesis.

Conclusion:
There is no statistically significant difference in sales between the regions.
```
